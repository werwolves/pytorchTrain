```python
import os
from glob import glob
import random
```

```python
cat_dir = r"cat_dog/cat"
dog_dir = r"cat_dog/dog"



def get_label(img_dir, label):
    train_path_l, var_path_l = [], []
    for i in os.listdir(img_dir) :
        if i.endswith(".jpg"):
            img_path = os.path.join(img_dir, i)
            if os.path.exists(img_path):
                if random.random() < 0.8:
                    train_path_l.append([f'{label}/{i}', label]) # [路径， 标签]
                else:
                    var_path_l.append([f'{label}/{i}', label])
    return train_path_l, var_path_l
                    
train1_l, val1_l = get_label(dog_dir, 'dog')
train2_l, val2_l = get_label(cat_dir, 'cat')

train_l = train1_l + train2_l
val_l = val1_l + val2_l

random.shuffle(train_l)
random.shuffle(val_l)

with open("train_list.txt", "w", encoding='utf-8') as train_writer:
    for i in train_l:
        train_writer.write(f"{i[0]}\t{i[1]}\n")
with open("val_list.txt", "w", encoding='utf-8') as var_writer:
    for i in val_l:
        var_writer.write(f"{i[0]}\t{i[1]}\n")                
                
                
                        
# 
#     
```
